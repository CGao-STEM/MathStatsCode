# 鞅

在下面的代码中，我们将分别模拟抛硬币中的fair game以及unfair game。

```python
import numpy as np
import numpy.random as nprd

T=5000
x0=0
def gen_martingale(p):
    X=[]
    x=x0
    for t in range(T):
        x=x+(1 if nprd.uniform()<p else -1)
        X.append(x)
    return X
        
import matplotlib.pyplot as plt
%matplotlib inline
# 设定图像大小
plt.rcParams['figure.figsize'] = (15.0, 8.0)
fig=plt.figure()
N=10
t=[t for t in range(T)]
corlors=np.linspace(0,1,N)
for i in range(N):
    x=gen_martingale(0.5)
    plt.plot(t,x,color=(0.8,1-corlors[i],corlors[i]))
plt.show() ## 画图
fig.savefig("martingale_coin5.pdf")
```

# 加倍赌注策略

```python
import numpy as np
import numpy.random as nprd

def gen_wealth(T):
    Y=[]
    y=1
    for i in range(T):
        c=(-2*y if y<0 else 1)
        epsilon=(1 if nprd.uniform()<0.5 else -1)
        y=y+c*epsilon
        Y.append(y)
    return Y
        
import matplotlib.pyplot as plt
%matplotlib inline
# 设定图像大小
plt.rcParams['figure.figsize'] = (15.0, 8.0)
fig=plt.figure()
N=20
T=300
t=[t for t in range(T)]
corlors=np.linspace(0,1,N)
for i in range(N):
    x=gen_wealth(T)
    plt.plot(t,x,color=(0.8,1-corlors[i],corlors[i]))
plt.show() ## 画图
fig.savefig("martingale_betting.pdf")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# 设定图像大小
plt.rcParams['figure.figsize'] = (15.0, 8.0)
fig=plt.figure()
T=80
N=1000000
PATH=[gen_wealth(T) for i in range(N)]
t=[t for t in range(T)]
m=[0 for t in range(T)]
for path in PATH:
    for tt in range(T):
        m[tt]+=path[tt]
m=[mm/N for mm in m]
plt.plot(t,m)
plt.show() ## 画图
fig.savefig("martingale_betting_average.pdf")
```

# 停止的上鞅的模拟

```python
import numpy as np
import numpy.random as nprd

T=5000
x0=100
def gen_martingale():
    X=[]
    x=x0
    for t in range(T):
        if x>=130:
            x=x
        else:
            x=x+(1 if nprd.uniform()<0.51 else -1)
        X.append(x)
    return X
        
import matplotlib.pyplot as plt
%matplotlib inline
# 设定图像大小
plt.rcParams['figure.figsize'] = (15.0, 8.0)
N=30
t=[t for t in range(T)]
corlors=np.linspace(0,1,N)
for i in range(N):
    x=gen_martingale()
    plt.plot(t,x,color=(0.8,1-corlors[i],corlors[i]))
plt.show() ## 画图
```

# 鞅的收敛

```python
import numpy as np
import numpy.random as nprd

T=2000
x0=1
def gen_martingale(p):
    X=[]
    x=x0
    for t in range(T):
        x*=(0.9 if nprd.uniform()<p else 1.1)
        X.append(x)
    return X
        
import matplotlib.pyplot as plt
%matplotlib inline
# 设定图像大小
plt.rcParams['figure.figsize'] = (5.0, 5.0)
fig=plt.figure()
N=8
t=[t for t in range(T)]
corlors=np.linspace(0,1,N)
for i in range(N):
    x=gen_martingale(0.5)
    plt.plot(t,x,color=(0.8,1-corlors[i],corlors[i]))
plt.show() ## 画图
fig.savefig("martingale_nonnegative_martingale5.pdf")
```
